```python
import matplotlib.pyplot as plt

# Define the color scheme and corresponding energy types
color_scheme = ['#014422', '#006727', '#248823', '#25A032', '#A4D279', '#D0E6B1']
energy_types = ['Other Renewables', 'Hydropower', 'Geothermal', 'Biofuels', 'Wind', 'Solar']

# Configure the figure
fig, ax = plt.subplots(1, figsize=(10, 2))

# Plot each color and label it with the corresponding energy type
for i, (color, energy) in enumerate(zip(color_scheme, energy_types)):
    ax.add_patch(plt.Rectangle((i, 0), 1, 1, color=color))
    ax.text(i + 0.5, -0.2, energy, ha='center', va='top', fontsize=10)

# Adjust layout and remove ticks
ax.set_xlim(0, len(color_scheme))
ax.set_ylim(-0.5, 1)
ax.set_yticks([])
ax.set_xticks([])
ax.set_title("Renewable Energy Colors by Type")

plt.tight_layout()
plt.show()
```
